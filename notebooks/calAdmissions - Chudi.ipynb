```python
#number of freshmen enrolled in each campus based on ethnicity
fresh_enrolled_by_campus = fresh.groupby(['Campus','Ethnicity']).enrolled.agg([sum()])
fresh_enrolled_by_campus.drop(index='All', level=1)
```

```python
#number of freshmen admitted in each campus organised by ethnicity
fresh_admitted_by_campus = fresh.groupby(['Campus','Ethnicity']).admitted.agg(['min', 'max'])
fresh_admitted_by_campus.drop(index='All', level=1)
```
