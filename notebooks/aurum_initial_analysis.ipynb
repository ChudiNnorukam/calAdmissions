```python
#Of the high schools that have students admitted, which high school has the highest ratio of admitted GPA to enrolled GPA? 
admGPA_to_enrGPA_ratio_dict = {}
count = 0
fresh_len = len(freshmen)
#fresh_len = 20

for index in range(fresh_len):
    #print()
    if (index % (fresh_len//20) == 0):
        print(count, "% of the way there!")
        count += 5
    
    high_school = freshmen["School"][index]
    
    adm_gpa = 0
    enr_gpa = 0
    
    if type(freshmen["adm (SUM)"][index]) == np.float:
        adm_gpa = int(freshmen["adm (SUM)"][index])
    
    if type(freshmen["enr (SUM)"][index]) == np.float:
        enr_gpa = int(freshmen["enr (SUM)"][index])
    
    #print ("adm_gpa:", adm_gpa, "enr_gpa:", enr_gpa)
    if (adm_gpa > 0): # if there are admitted students from the high school
        if (enr_gpa > 0): # if there are enrolled students in the high school
            #print ("ratio for" , freshmen["School"][index], ":", round((adm_gpa/enr_gpa), 5))
            admGPA_to_enrGPA_ratio_dict[high_school] = (adm_gpa)/(enr_gpa) #find the adm_gpa/enr_gpa ratio
#        else: # no enrolled students
#            a = 1 # do nothing
#            #admGPA_to_enrGPA_ratio_dict[high_school] = 0 #set the ratio to 0
#    else: # no admit students
#        a = 1 # do nothing
#        #admGPA_to_enrGPA_ratio_dict[high_school] =  0 #np.nan # set the ratio to np.nan

print (admGPA_to_enrGPA_ratio_dict)
```
