```python
trans = pd.read_csv("Transfers.csv")
trans
```

```python
#How is the ratio of admitted students to applied students differ amongst different ethnicities at UC Berkeley Over various Semesters?
#--Data Frames only cover Fall 2017--
```

```python
trans = trans.rename(columns = {'adm (SUM)':'admitted', 'app (SUM)':'applied', 'enr (SUM)':'enrolled', 'Uad Uc Ethn 6 Cat':'Ethnicity'})
```

```python
fresh = fresh.rename(columns = {'adm (SUM)':'admitted', 'app (SUM)':'applied', 'enr (SUM)':'enrolled', 'Uad Uc Ethn 6 Cat':'Ethnicity'})
```

```python
gpa = gpa.rename(columns = {'adm (SUM)':'admitted_gpa', 'app (SUM)':'applied_gpa', 'enr (SUM)':'enrolled_gpa'})
```

```python
trans_2017 = trans[trans["Fall Term"] == 2017]
```

```python
fresh_2017 = fresh[fresh["Fall Term"] == 2017]
```

```python
fresh_2017
```

```python
#number of freshmen that applied to each campus based on ethnicity in Fall 2017
fresh_applied_by_campus = fresh_2017.groupby(['Campus','Ethnicity']).applied.agg(['sum'])
fresh_applied_by_campus #.drop(index='All', level=1)
```

```python
#number of freshmen admitted to each campus based on ethnicity in Fall 2017
fresh_admitted_by_campus = fresh_2017.groupby(['Campus','Ethnicity']).admitted.agg(['sum'])
fresh_admitted_by_campus
```

```python
#number of transfer students that applied to each campus organised by ethnicity in Fall 2017
trans_applied_by_campus = trans_2017.groupby(['Campus','Ethnicity']).applied.agg(['sum'])
trans_applied_by_campus #.drop(index='All', level=1)
```

```python
#number of transfer students admitted to each campus, organised by ethnicity in Fall 2017
trans_admitted_by_campus = trans_2017.groupby(['Campus','Ethnicity']).admitted.agg(['sum'])
trans_admitted_by_campus #.drop(index='All', level=1)
```
