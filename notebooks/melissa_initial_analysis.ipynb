```python
#grouping each data table by 'Campus' and sorting them based on different admitted values
```

```python
freshmen_by_campus = freshmen.groupby(["Campus"]).count().sort_values(by="adm (SUM)")
freshmen_by_campus
```

```python
freshmen_by_campus_drop = freshmen_by_campus.drop(['Calculation1', 'City', 'County/State/ Territory', 'Fall Term', 'School', 'Sch Src FR', 'Uad Uc Ethn 6 Cat'], axis=1)
freshmen_by_campus_drop.plot.bar()
```

```python
#proportion of freshmen admitted into the university system compared with the number that applied
prop_fresh = freshmen_by_campus['adm (SUM)'] / freshmen_by_campus['app (SUM)']
prop_fresh
```

```python
prop_fresh.sort_values(ascending=False).plot.line(rot = 90)
```

```python
sorted_gpa = gpa.groupby(["Campus"]).mean().sort_values(by="Adm GPA (SUM)")
```

```python
sorted_gpa_drop = sorted_gpa.drop(['Fall Term'], axis=1)
sorted_gpa_drop.plot.bar()
```

```python
transfers_by_campus = transfers.groupby(["Campus"]).sum().sort_values(by="adm (SUM)")
transfers_by_campus
```

```python
transfers_by_campus_drop = transfers_by_campus.drop(['Fall Term'], axis=1)
transfers_by_campus_drop.plot.bar()
```

```python
transfers_by_campus.iloc[0].values
```

```python
prop_transfers = transfers_by_campus['adm (SUM)'] / transfers_by_campus['app (SUM)']
prop_transfers
```

```python
#proportion of transfers admitted into the university system compared with the number that applied
prop_transfers.sort_values(ascending=False).plot.line(rot = 90)
```
