```python
#import database regarding wealth distributions from different counties in 2016
county_wealth = pd.read_csv("../data/CA County Wealth Data.csv")
county_wealth.head(30)
```

```python
#create a list of schools that had students accepted into Cal for the Fall 2018 Term

count = 0
fresh_len = len(freshmen)
#fresh_len = 100
i_list = []

for i in range(fresh_len):
    if (i % (fresh_len//20) == 0):
        print(count, "% of the way there!")
        count += 5
    if (freshmen.iloc[i][1] == "Berkeley") and (freshmen.iloc[i][4] == 2016):
        #print ("FINALLY")
        i_list.append(i)    
        
berk_acceptance_2016_schools_list = [freshmen.iloc[i] for i in i_list]
```

```python
#create a copy of the wealth database with only data from 2016 and excluding the 'geo' column
county_2016 = county_wealth.iloc[:, [0, 1, 3]][county_wealth['year'].isin([2016, '2016'])] 
#print ("county_2016: \n", county_2016)

#create a copy of the 2016 database that removes the discrepancies between county names
short_county_2016 = county_2016[:]
for i in range(len(county_2016)):
    index = county_2016.iloc[i, 1].find("County") 
    #print (county_2016.iloc[i, 1])
    if (index != -1):
        short_county_2016.iloc[i, 1] = county_2016.iloc[i, 1][:index].strip()
    else:
        short_county_2016.iloc[i, 1] = county_2016.iloc[i, 1][:].strip()
    #print (short_county_2016.iloc[i, 1])
#print (short_county_2016)

#transform this into a dictionary formatted like this:
#   {county: income}

county_income_dict = {}
for i in range(len(short_county_2016)):
    county_income_dict[short_county_2016.iloc[i, 1]] = short_county_2016.iloc[i, 2]

print(county_income_dict)
```

```python
import math

for i in range(len(berk_acceptance_2016_schools_list)):
    if (not(math.isnan(berk_acceptance_2016_schools_list[i]["enr (SUM)"]))):
        print("{school:35s} -> applied:{applied:4.0f}  /  admitted:{admitted:3.0f}  /  enrolled:{enrollees:3.0f}".format(
            school = berk_acceptance_2016_schools_list[i]["School"], 
            applied = berk_acceptance_2016_schools_list[i]["app (SUM)"],
            admitted = berk_acceptance_2016_schools_list[i]["adm (SUM)"],
            enrollees = berk_acceptance_2016_schools_list[i]["enr (SUM)"]
        ))
```

```python
#ca_pu_2016_berk is a dataframe that holds acceptance numbers for schools that have students who enrolled in the 
# 2016 fall term to berkeley. 

# Assumptions: 
#  1)a school with Nan data for admittance is assumed to be the maximum admitted, 3.
#  2)a school with Nan data for enrollment is assumed to be 2, a general upper bound for the assumed yield of Berkeley

 #Column 0 -> County/State/Territory   -> County of Origin for the high school
 #Column 1 -> School                   -> School Name
 #Column 2 -> adm (SUM)                -> number of students from the school admitted
 #Column 3 -> app (SUM)                -> number of students from the school who applied
 #Column 4 -> enr (SUM)                -> number of students from the school who enrolled

ca_pub_2016_berk = freshmen.iloc[:, [3, 5, 8, 9, 10]][freshmen['Sch Src FR'].isin(['California Public High School'])][freshmen['Campus'].isin(['Berkeley'])][freshmen['enr (SUM)'] > 0]

print(ca_pub_2016_berk)
```

```python
#Create a DataFrame that takes in the counts of application, admittance, and enrollment for all schools in a given county,
# and then concatenates data from the wealth data of the county to draw socioeconomic conclusions about admissions rate

#format for dictionary county_dict => {'County': [Applied, Admitted, Enrolled, Income]}

#county is an object in the ca_pub_2016_berk dictionary

county_dict = {}
for i in range(len(ca_pub_2016_berk)):
    county = ca_pub_2016_berk.iloc[i]
    county_name = county[0]
    #create the entry in the dict for a county if it doesn't already exist
    if (county_name not in county_dict):
        county_dict[county_name] = [0, 0, 0, county_income_dict[county_name]]
    #update the entry for the county with the school's data
    num_applied, num_admitted, num_enrolled = ca_pub_2016_berk.iloc[i][3], ca_pub_2016_berk.iloc[i][2], ca_pub_2016_berk.iloc[i][4]
    county_num_applied, county_num_admitted, county_num_enrolled = county_dict[county_name][0], county_dict[county_name][1], county_dict[county_name][2]
    county_dict[county_name][0] = county_num_applied  + num_applied
    county_dict[county_name][1] = county_num_admitted + num_admitted
    county_dict[county_name][2] = county_num_enrolled + num_enrolled        

      
county_admissions_income = pd.DataFrame(data=county_dict)
print(county_admissions_income)
print('a')
```
